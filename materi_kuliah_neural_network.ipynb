Fungsi sigmoid adalah fungsi aktivasi. Untuk sembarang angka input n, fungsi sigmoid memetakan angka tersebut ke output antara 0 dan 1.
Ketika nilai n semakin besar, nilai output semakin mendekati 1 dan ketika n semakin kecil, nilai output semakin mendekati 0.

```python
def sigmoid(input):    
    output = 1 / (1 + np.exp(-input))
    return output
```

fungsi turunan dari fungsi sigmoid

```python
def sigmoid_derivative(input):
    return sigmoid(input) * (1.0 - sigmoid(input))
```
